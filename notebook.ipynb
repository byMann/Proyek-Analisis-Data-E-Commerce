### Cleaning Data

```python
# Fix data types

# convert timestamp to datetime
order_items_df['shipping_limit_date'] = pd.to_datetime(order_items_df['shipping_limit_date'])

datetime_columns = ["order_purchase_timestamp", "order_approved_at", "order_delivered_carrier_date", "order_delivered_customer_date", "order_estimated_delivery_date"]

for column in datetime_columns:
  orders_df[column] = pd.to_datetime(orders_df[column])

orders_df.head()

# checking seller city name is alphabet
is_numeric = sellers_df['seller_city'].str.isnumeric()

if is_numeric.any():
    state = sellers_df['seller_state']
    sellers_df.loc[is_numeric, 'seller_city'] = state + " city"
```

```python
# Missing value

# Dropping purchased orders that's not approved by seller
orders_df['order_purchase_timestamp'].isnull().sum()
orders_df['order_approved_at'].isnull().sum()

orders_df = orders_df.dropna(subset=['order_approved_at'])

# Dropping null product category 
products_df['product_category_name'].isnull().sum()
products_df.dropna(subset=['product_category_name'], inplace=True)
```

**Insight:**
- Converting timestamp to avoid misleading to incorrect sorting or filtering results.
- Checks for missing values and removes all entries with a null value, ensure valid data for analysis

## Exploratory Data Analysis (EDA)

### Explore ...

```python
customers_df.customer_id.is_unique
```

```python
orders_customers_df = pd.merge(
    left=orders_df,
    right=customers_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
orders_customers_df.head()
```

```python
result_df = pd.merge(customers_df, orders_df, left_on='customer_id', right_on='customer_id', how='inner')

# Gabungkan hasil dengan order_items menggunakan order_id
result_df = pd.merge(result_df, order_items_df, left_on='order_id', right_on='order_id', how='inner')

# Gabungkan hasil dengan sellers menggunakan seller_id
result_df = pd.merge(result_df, sellers_df, left_on='seller_id', right_on='seller_id', how='inner')

# Gabungkan hasil dengan products menggunakan product_id
result_df = pd.merge(result_df, products_df, left_on='product_id', right_on='product_id', how='inner')

# Gabungkan hasil dengan product_category_name_translation menggunakan product_category_name
result_df = pd.merge(result_df, product_category_name_df, left_on='product_category_name', right_on='product_category_name', how='inner')

# Cetak hasil gabungan data
result_df.to_csv('main_data.csv', index=False)
result_df.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: How do product categories contribute to the total imports of the countries in 2017?

```python
impor = result_df[result_df['customer_state'] != result_df['seller_state']]
impor = impor[['order_id', 'customer_state', 'seller_state', 'product_category_name', 'order_purchase_timestamp', 'product_category_name_english']]

impor['year'] = impor['order_purchase_timestamp'].dt.year

total_sales_import = (
    impor[impor['year'] == 2017]  # Filter untuk tahun 2017
    .groupby(['product_category_name_english'])['order_id']
    .count()
    .reset_index()
)

total_sales_import.rename(columns={'order_id': 'total_sales'}, inplace=True)

top_import_categories = total_sales_import.nlargest(5, 'total_sales')

plt.figure(figsize=(12, 8)) 

bars = sns.barplot(data=top_import_categories, 
                   x='product_category_name_english', 
                   y='total_sales', 
                   palette='Set2')

plt.xticks([])
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.title('Top 5 Import Product Categories in 2017')

plt.legend(title='Product Categories', 
           labels=top_import_categories['product_category_name_english'], 
           loc='upper right', 
           handlelength=2,
           fontsize='medium')

plt.tight_layout()
plt.show()
```

### Pertanyaan 2: How is the distribution of sellers from each country and city?

```python
sellersbystates_df = result_df.groupby(by="seller_state").agg({"seller_id": "nunique"}).reset_index()
sellersbystates_df.rename(columns={"seller_id": "seller_count"}, inplace=True)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 10))

colors_state = ["#90CAF9"] + ["#D3D3D3"] * (len(sellersbystates_df) - 1)

sns.barplot(
    x="seller_count", 
    y="seller_state", 
    data=sellersbystates_df.sort_values(by="seller_count", ascending=False).head(10),
    palette=colors_state,
    ax=ax1,
    hue="seller_state",
    legend=False
)
ax1.set_title("Number of Sellers by State", loc="center", fontsize=30)
ax1.set_ylabel(None)
ax1.set_xlabel(None)
ax1.tick_params(axis='x', labelsize=15)
ax1.tick_params(axis='y', labelsize=20)
sellers_city_df = result_df.groupby(by="seller_city").agg({"seller_id": "nunique"}).reset_index()
sellers_city_df.rename(columns={"seller_id": "seller_count"}, inplace=True)

sellers_city_df = sellers_city_df.sort_values(by="seller_count", ascending=False).head(10)

colors_city = ["#90CAF9"]
colors_city += ["#D3D3D3"] * (len(sellers_city_df) - 1)

sns.barplot(
    x="seller_count", 
    y="seller_city", 
    data=sellers_city_df,
    palette=colors_city,
    ax=ax2,
    hue="seller_city",
    legend=False
)

ax2.set_title("Number of Sellers by City", loc="center", fontsize=30)
ax2.set_ylabel(None)
ax2.set_xlabel(None)
ax2.tick_params(axis='x', labelsize=15)
ax2.tick_params(axis='y', labelsize=20)

plt.tight_layout()
plt.show()
```

**Insight:**
- Biggest Product Categories imported by customers is `bed_bath_table`
- State with largest sellers is `SP` and City with largest sellers is `sao paulo`

## Analisis Lanjutan (Opsional)

### Customer Behaviour Based on RFM Parameters

```python
# rfm_df = pd.merge(customers_df, orders_df, on='customer_id', how='inner')
# rfm_df = pd.merge(rfm_df, order_items_df, on='order_id', how='inner')

rfm_df = result_df.groupby(by="customer_state", as_index=False).agg({
    "order_purchase_timestamp": "max",
    "order_id": "nunique",
    "price": "sum"
})

rfm_df.columns = ["customer_state", "last_purchase_date", "frequency", "monetary"]
rfm_df["last_purchase_date"] = pd.to_datetime(rfm_df["last_purchase_date"]).dt.date

recent_date = orders_df["order_purchase_timestamp"].dt.date.max()
rfm_df["recency"] = rfm_df["last_purchase_date"].apply(lambda x: (recent_date - x).days)

rfm_df.drop("last_purchase_date", axis=1, inplace=True)

colors = sns.color_palette("Blues")

plt.figure(figsize=(18, 5))

rfm_df = rfm_df.dropna(subset=['customer_state', 'recency', 'frequency', 'monetary'])
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(35, 15))

# Plot for Recency
sns.barplot(y="recency", x="customer_state", data=rfm_df.sort_values(by="recency", ascending=True).head(5), ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel("Customer State", fontsize=30)
ax[0].set_title("By Recency (days)", loc="center", fontsize=50)
ax[0].tick_params(axis='y', labelsize=30)
ax[0].tick_params(axis='x', labelsize=30) 
ax[0].set_xticks(range(5))
ax[0].set_xticklabels(ax[0].get_xticklabels(), ha='right', fontsize=30)

# Plot for Frequency
sns.barplot(y="frequency", x="customer_state", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel("Customer State", fontsize=30)
ax[1].set_title("By Frequency", loc="center", fontsize=50)
ax[1].tick_params(axis='y', labelsize=30)
ax[1].tick_params(axis='x', labelsize=30)
ax[1].set_xticks(range(5))
ax[1].set_xticklabels(ax[1].get_xticklabels(), ha='right', fontsize=30)

# Plot for Monetary
sns.barplot(y="monetary", x="customer_state", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel("Customer State", fontsize=30)
ax[2].set_title("By Monetary", loc="center", fontsize=50)
ax[2].tick_params(axis='y', labelsize=30)
ax[2].tick_params(axis='x', labelsize=30)
ax[2].set_xticks(range(5))
ax[2].set_xticklabels(ax[2].get_xticklabels(), ha='right', fontsize=30)

plt.tight_layout()
plt.show()
```

**Insight:**
RFM analysis focusing on segmenting customers in order to achieve more profit by analyzing customer's behaviour (in period of time)
